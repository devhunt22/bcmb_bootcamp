5. Write a for loop to test if each flower measurement is a normal distribution. If the distribution is normal print the p-value, if the distribution is not normal print "Distribution not normal!" (hint scipy has a normaltest function https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.normaltest.html)

```python
#Each flower variety has four measurements associated with it 
    #petal.length, petal.width, sepal.length, sepal.width 
    #Need to pull out each category as a header 
measure_list=['petal.length', 'petal.width', 'sepal.length', 'sepal.width']
print(measure_list)
```

```python
#Need to generate a for loop that will evaluate all the values associated with each measurement in a given variety 
for i in measure_list:
    array=np.array(iris_raw[i])
    print(array)
```

```python
for i in measure_list:
    array=np.array(iris_raw[i])
    #print(array) #This will print all the numbers in the array associated with each measure 
    x=stats.normaltest(array)
    print(x)
    print(x.pvalue)
```

```python
#Need to generate a for loop that will evaluate all the values associated with each measurement in a given variety 
for i in measure_list:
    array=np.array(iris_raw[i])
    #print(array) #This will print all the numbers in the array associated with each measure 
    x=stats.normaltest(array) #Define a new variable 'x' as the boolean (T/F) for array normal dist.
    pval=x.pvalue #defined new variable pval as the value of p for the normal dist on array
    if pval < 0.05: #if the value of p is less than 0.05, return the value 
        print(pval)
    else:
        print('Distribution is not normal.') #else, say the dist is not normal 
```
